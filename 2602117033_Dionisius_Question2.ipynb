```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
df = pd.read_csv("UAS_LAB_AIDS_Classification_Dataset.csv")
df
```

```python
#Check Missing Value
df.isnull().sum().sort_values(ascending=True)
```

```python
#Data Modelling
x = df.drop('infected', axis=1)
y = df['infected']

#Split Data (Ratio 80:20)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
from sklearn.pipeline import make_pipeline
import warnings
warnings.filterwarnings('ignore')
```

```python
#1b.
from sklearn.tree import DecisionTreeClassifier
singleTree_model = DecisionTreeClassifier()
singleTree_model.fit(x_train, y_train)
y_singleTree = singleTree_model.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_singleTree))
```

```python
#1c.
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
import warnings
warnings.filterwarnings('ignore')
baggingTree_model = DecisionTreeClassifier()
baggingmodel = BaggingClassifier(base_estimator=baggingTree_model, n_estimators=25)
baggingmodel.fit(x_train,y_train)
y_baggingTree = baggingmodel.predict(x_test)
y_baggingTree
```

```python
#1c.
from sklearn.metrics import classification_report
print(classification_report(y_test,y_baggingTree, target_names = ['0','1']))
```
